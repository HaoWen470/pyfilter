# Nutria
In this Notebook we'll consider the population growth of the Nutria species. The data has been taken from .. . We'll begin importing the data and visualizing it.

```python
import pandas as pd

data = pd.read_csv('nutria.txt', sep='\t').iloc[:, 0].rename('nutria')

data.plot(figsize=(16, 9), title='Nutria population')
```

Next, we'll specify the model to use for inference. We'll use the flexible Allee model, found in .. . However, instead of considering the actual population, we'll use the logarithm.

```python
%cd ..

from pyfilter.timeseries import LinearGaussianObservations, AffineModel
from torch.distributions import Normal, Gamma, TransformedDistribution, AffineTransform, PowerTransform
import torch

def f(x, a, b, c, d):
    exped = x.exp()
    return x + a + b * exped + c * exped ** 2


def g(x, a, b, c, d):
    return d.sqrt()


def f0(a, b, c, d):
    return torch.tensor(0., device=a.device)


def g0(a, b, c, d):
    return d.sqrt()


alpha = data.shape[0] / 2
beta = 2 * (alpha - 1) / 10

invgamma = TransformedDistribution(Gamma(alpha, beta), PowerTransform(torch.tensor(-1.)))

h_priors = Normal(0., 1.), Normal(0., 1.), Normal(0., 1.), invgamma
noise = Normal(0., 1.), Normal(0., 1.)
hidden = AffineModel((f0, g0), (f, g), h_priors, noise)

model = LinearGaussianObservations(hidden, 1., invgamma)
```

Next, we'll use SMC2 together with APF to perform inference on the logged dataset.

```python
from pyfilter.algorithms import SMC2
from pyfilter.filters import SISR
from pyfilter.resampling import multinomial
import numpy as np


logged_data = torch.from_numpy(np.log(data).values)

algs = list()
for i in range(2):
    filt = SISR(model.copy(), 300, resampling=multinomial)
    alg = SMC2(filt, 1000, threshold=0.5).initialize().fit(logged_data)
    
    algs.append(alg)
```

Next, let's visualize the filtered means of the state.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16, 9))

data.plot(ax=ax)

for alg in algs:
    ax.plot(alg.filter.filtermeans.mean(dim=1).exp().cpu().numpy(), label='Filtered')

ax.legend()
```

Next, let's visualize the posterior distributions of the parameters.

```python
from pyfilter.utils import normalize
import pandas as pd

fig, ax = plt.subplots(len(alg.filter.ssm.flat_theta_dists), figsize=(16, 9))

for alg in algs:
    w = normalize(alg._w_rec)
    for i, param in enumerate(alg.filter.ssm.flat_theta_dists):    
        xrange, xvals = param.get_plottable(weights=w)

        ax[i].plot(xrange, xvals)
```
